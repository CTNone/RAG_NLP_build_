# Cào dữ liệu


```python

import requests
from bs4 import BeautifulSoup
import pdfplumber
import json
import os
from urllib.parse import urljoin
import re

# Thư mục lưu trữ dữ liệu
OUTPUT_DIR = "pittsburgh_data"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Hàm cào dữ liệu HTML
def scrape_html(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')

        # Lấy tiêu đề trang
        title = soup.title.string if soup.title else "No Title"

        # Loại bỏ script, style, nav, footer, header
        for element in soup(["script", "style", "nav", "footer", "header"]):
            element.decompose()

        content = soup.get_text(separator='\n', strip=True)
        return {"url": url, "title": title, "content": content}
    except Exception as e:
        print(f"Lỗi khi cào HTML {url}: {e}")
        return None

# Hàm trích xuất văn bản từ PDF
def extract_pdf(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        pdf_path = os.path.join(OUTPUT_DIR, "temp.pdf")

        with open(pdf_path, "wb") as f:
            f.write(response.content)

        with pdfplumber.open(pdf_path) as pdf:
            text = ""
            for page in pdf.pages:
                text += page.extract_text() or ""

        os.remove(pdf_path)
        return {"url": url, "title": "PDF Document", "content": text}
    except Exception as e:
        print(f"Lỗi khi xử lý PDF {url}: {e}")
        return None

# Hàm tìm liên kết PDF từ trang web
def find_pdf_links(url, base_url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        pdf_links = []

        for a in soup.find_all('a', href=True):
            href = a['href']
            if href.lower().endswith('.pdf'):
                pdf_url = urljoin(base_url, href)
                pdf_links.append(pdf_url)

        return pdf_links
    except Exception as e:
        print(f"Lỗi khi tìm PDF tại {url}: {e}")
        return []

# Hàm lưu dữ liệu
def save_data(data, index):
    # Tạo tên tệp từ URL hoặc index
    safe_url = re.sub(r'[^\w\-]', '_', data["url"])[:50]
    output_file = os.path.join(OUTPUT_DIR, f"doc_{index}_{safe_url}.json")
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

# Hàm chính
def scrape_urls(url_list):
    all_data = []
    index = 1

    for url in url_list:
        print(f"Đang xử lý: {url}")

        # Xử lý HTML
        if not url.endswith('.pdf'):
            data = scrape_html(url)
            if data:
                save_data(data, index)
                all_data.append(data)
                index += 1

            # Tìm và xử lý PDF
            pdf_links = find_pdf_links(url, url)
            for pdf_url in pdf_links:
                pdf_data = extract_pdf(pdf_url)
                if pdf_data:
                    save_data(pdf_data, index)
                    all_data.append(pdf_data)
                    index += 1
        else:
            # Xử lý trực tiếp PDF
            data = extract_pdf(url)
            if data:
                save_data(data, index)
                all_data.append(data)
                index += 1

    # Lưu danh sách URL
    with open(os.path.join(OUTPUT_DIR, "url_list.json"), 'w', encoding='utf-8') as f:
        json.dump(url_list, f, ensure_ascii=False, indent=2)

    print(f"Hoàn thành! Dữ liệu được lưu tại: {OUTPUT_DIR}")
    print(f"Tổng số tài liệu: {len(all_data)}")

# Danh sách URL (sẽ được cung cấp bên dưới)
URL_LIST = [
    # Thêm danh sách URL tại đây
]

URL_LIST = [
    # Wikipedia - Pittsburgh Tổng quan
    "https://en.wikipedia.org/wiki/Pittsburgh",
    "https://en.wikipedia.org/wiki/History_of_Pittsburgh",
    "https://en.wikipedia.org/wiki/Culture_of_Pittsburgh",
    "https://en.wikipedia.org/wiki/Sports_in_Pittsburgh",
    "https://en.wikipedia.org/wiki/List_of_Pittsburgh_neighborhoods",
    "https://en.wikipedia.org/wiki/Pittsburgh_Public_Schools",
    "https://en.wikipedia.org/wiki/Economy_of_Pittsburgh",

    "https://en.wikipedia.org/wiki/Pittsburgh_Steelers",
    "https://en.wikipedia.org/wiki/Pittsburgh_Pirates",
    "https://en.wikipedia.org/wiki/Pittsburgh_synagogue_shooting",
    "https://en.wikipedia.org/wiki/Pittsburgh_Penguins",
    "https://en.wikipedia.org/wiki/Pittsburgh_International_Airport",
    "https://en.wikipedia.org/wiki/Pittsburgh_Post-Gazette",
    "https://en.wikipedia.org/wiki/Pittsburgh_Marathon",
    "https://en.wikipedia.org/wiki/University_of_Pittsburgh",
    "https://en.wikipedia.org/wiki/Pittsburgh_Regional_Transit",
"""CMU wiki"""
    "https://en.wikipedia.org/wiki/Carnegie_Mellon_University",



    # Carnegie Mellon University (CMU) - Sự kiện, Giáo dục
    "https://www.cmu.edu/events/",
    "https://www.cmu.edu/about/history.html"


    # Pittsburgh City Paper - Văn hóa, Ẩm thực, Sự kiện
    "https://www.pittsburghpa.gov/Events-directory"
    "https://www.pittsburghcitypaper.com/pittsburgh/arts-entertainment/",
    "https://www.pittsburghcitypaper.com/pittsburgh/food-drink/",
    "https://www.pittsburghcitypaper.com/pittsburgh/events/",
    "https://www.pittsburghcitypaper.com/pittsburgh/news/",

    # Trang chính thức của thành phố Pittsburgh - Ngân sách, Thuế, Dịch vụ
    "https://pittsburghpa.gov/finance/budget",
    "https://pittsburghpa.gov/finance/tax-descriptions",
    "https://pittsburghpa.gov/mayor/history",
    "https://pittsburghpa.gov/publicsafety/",
    "https://pittsburghpa.gov/city-info/",

    # Các nguồn ẩm thực
    "https://www.allrecipes.com/what-is-pittsburgh-salad-8679005",
    "https://www.pittsburghmagazine.com/best-restaurants/",
    "https://www.goodfoodpittsburgh.com/",

    # Các nguồn thể thao

    "https://www.steelers.com/",
    "https://www.nhl.com/penguins",

    # Các nguồn văn hóa và lịch sử
    "https://www.heinzhistorycenter.org/",
    "https://www.carnegiemuseums.org/",
    "https://www.frickpittsburgh.org/",

    # Giáo dục và trường đại học
    "https://www.pitt.edu/about/history",
    "https://www.duq.edu/about/history",
    "https://www.chatham.edu/about/history",

    # Sự kiện và lễ hội
    "https://www.pgharts.org/",
    "https://www.threeriversartsfestival.org/",
    "https://www.pittsburghpride.org/",

    # Kinh tế và việc làm
    "https://www.pittsburghregion.org/economy/",
    "https://www.pghworks.com/",

    # Khí hậu và môi trường
    "https://www.weather.gov/pbz/climate",
    "https://pittsburghpa.gov/sustainability/",

    # Khu phố và cộng đồng
    "https://www.pittsburghbeautiful.com/neighborhoods/",
    "https://www.lawrencevillepgh.com/",
    "https://southsidepgh.com/",

    # Tội phạm và an toàn
    "https://www.pittsburghpa.gov/police/statistics",
    "https://www.neighborhoodscout.com/pa/pittsburgh/crime",

    # Các nguồn bổ sung
    "https://www.pittsburghzoo.org/",
    "https://www.phipps.conservatory.org/",
    "https://www.post-gazette.com/local/pittsburgh-history",
    "https://www.discovertheburgh.com/"
]
if __name__ == "__main__":
    scrape_urls(URL_LIST)
```

```python
import json
import re
import os
import unicodedata

# Đường dẫn thư mục đầu vào và đầu ra
INPUT_DIR = "pittsburgh_data"
OUTPUT_DIR = "pittsburgh_data/cleaned"

# Tạo thư mục đầu ra nếu chưa tồn tại
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Hàm làm sạch văn bản
def clean_text(text):
    # Chuẩn hóa Unicode về dạng NFC (đảm bảo UTF-8 nhất quán)
    text = unicodedata.normalize('NFC', text)

    # Loại bỏ ký tự đặc biệt không cần thiết (giữ lại chữ cái, số, dấu câu cơ bản)
    text = re.sub(r'[^\w\s.,!?()-]', ' ', text)

    # Loại bỏ các đoạn quảng cáo hoặc thông tin không liên quan của Wikipedia
    text = re.sub(r'Jump to content|Coordinates:.*?\n|From Wikipedia, the free encyclopedia|This article is about.*?For other uses,.*?disambiguation\)|Not to be confused.*?\n', '', text, flags=re.DOTALL)

    # Loại bỏ các liên kết nội bộ hoặc chú thích của Wikipedia
    text = re.sub(r'\[\d+\]|\[edit\]|See also:.*?\n', '', text)

    # Loại bỏ khoảng trắng dư thừa và dòng trống
    text = re.sub(r'\n+', '\n', text)
    text = re.sub(r'\s+', ' ', text.strip())

    # Chuẩn hóa dấu câu
    text = re.sub(r'\s+([.,!?])', r'\1', text)

    return text

# Hàm làm sạch một tệp JSON
def clean_json_file(input_file, output_file):
    try:
        # Đọc tệp JSON
        with open(input_file, 'r', encoding='utf-8') as f:
            data = json.load(f)

        # Làm sạch trường content
        if 'content' in data:
            data['content'] = clean_text(data['content'])

        # Lưu tệp JSON đã làm sạch
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=2)

        print(f"Đã làm sạch và lưu tệp tại: {output_file}")
        return True
    except Exception as e:
        print(f"Lỗi khi xử lý tệp {input_file}: {e}")
        return False

# Hàm làm sạch tất cả tệp JSON trong thư mục
def clean_all_files(input_dir, output_dir):
    processed_files = 0
    failed_files = 0

    for filename in os.listdir(input_dir):
        if filename.endswith('.json') and not filename.startswith('url_list'):
            input_file = os.path.join(input_dir, filename)
            output_file = os.path.join(output_dir, filename.replace('.json', '_cleaned.json'))

            if clean_json_file(input_file, output_file):
                processed_files += 1
            else:
                failed_files += 1

    print(f"Hoàn thành! Đã xử lý {processed_files} tệp, lỗi {failed_files} tệp.")
    print(f"Kết quả được lưu tại: {output_dir}")

# Chạy hàm chính
if __name__ == "__main__":
    clean_all_files(INPUT_DIR, OUTPUT_DIR)
```
